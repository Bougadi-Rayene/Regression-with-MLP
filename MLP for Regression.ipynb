```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import tensorflow as tf
```

```python
sigma = 0.2
eps = np.random.normal(0, sigma,  size= 5000)
p = 0.1
```

```python
x = np.linspace(-10,
               10,
               num = 5000,
               endpoint = True,
               retstep = False,
               dtype = None)
```

```python
y = p * x * np.cos(x)+ eps
```

```python
len(x)
```

```python
plt.figure(figsize=(12 , 6))
plt.scatter( x , y , c='b')
```

```python
'''x_train = x[:4000]
y_train = y[:4000]

x_test = x[4000:]
y_test= y[4000:]  '''
```

```python
x = x.reshape(-1, 1)
y = y.reshape(-1, 1)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1,)),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1)  # une seule couche de sortie pour la régression (sans activation : activation lineaire)
])

model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
history = model.fit(x, y, epochs=100, verbose=0)
```

```python
y_predict = model.predict(x)

plt.figure(figsize=(12 , 6))
plt.scatter( x , y , c='b')
plt.plot( x , y_predict , c='r')
```

```python
test_loss = model.evaluate(x , y)
```

```python
plt.plot(history.history['loss'])   #training loss
plt.xlabel('Epochs')
plt.ylabel('Loss')
```
